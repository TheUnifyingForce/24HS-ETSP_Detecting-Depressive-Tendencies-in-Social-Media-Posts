## Error Heatmap

### 2 Classes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

misclassification_data = {
    '1 roberta sentiment': [5, 13],
    '4 fine-tuned sent c3': [6, 11],
    '5 roberta base': [11, 12],
    '8 fine-tuned base c3': [9, 15]
}

labels = ['0: healthy', '1: depressed']
df = pd.DataFrame(misclassification_data, index=labels)

plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap="Blues", fmt='.0f', cbar=True, linewidths=.5)

plt.title("2 Classes Error Heatmap", fontsize=16)
plt.xlabel("Models", fontsize=12)
plt.ylabel("Labels", fontsize=12)

plt.show()
```
